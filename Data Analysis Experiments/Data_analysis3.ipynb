```python
import re

def convert_height_to_inches(height):
    if pd.isna(height):  
        return None
    match = re.match(r"(\d+)ft *(\d+)?in?", str(height))
    if match:
        feet = int(match.group(1))
        inches = int(match.group(2)) if match.group(2) else 0
        return feet * 12 + inches
    return None

def replace_nans(dataset):
    if 'waist' in dataset.columns:
        mode_waist = dataset['waist'].mode()[0]
        dataset['waist'] = dataset['waist'].fillna(mode_waist)
        print(f"Replaced NaN in 'waist' with mode: {mode_waist}")

    if 'size' in dataset.columns:
        mode_size = dataset['size'].mode()[0]
        dataset['size'] = dataset['size'].fillna(mode_size)
        print(f"Replaced NaN in 'size' with mode: {mode_size}")

    if 'height' in dataset.columns:
        dataset['height_in_inches'] = dataset['height'].apply(convert_height_to_inches)
        median_height = dataset['height_in_inches'].median()
        dataset['height_in_inches'] = dataset['height_in_inches'].fillna(median_height)
        print(f"Replaced NaN in 'height' with median (in inches): {median_height}")
        dataset['height'] = dataset['height_in_inches']  
        dataset.drop(columns=['height_in_inches'], inplace=True)

    if 'age' in dataset.columns:
        median_age = dataset['age'].median()  
        dataset['age'] = dataset['age'].fillna(median_age)
        print(f"Replaced NaN in 'age' with median: {median_age}")

    return dataset

cleaned_dataset_modcloth_df = replace_nans(modcloth_df)
```

```python
def compute_statistics(data, dataset_name):

    # Basic stats
    num_transactions = len(data)
    num_customers = data['user_id'].nunique()
    num_products = data['item_id'].nunique()

    # Additional stats
    avg_transactions_per_customer = num_transactions / num_customers
    avg_transactions_per_product = num_transactions / num_products
    median_transactions_per_customer = data.groupby('user_id').size().median()
    median_transactions_per_product = data.groupby('item_id').size().median()
    
    # Fit feedback ratios
    fit_feedback_ratios = data['fit'].value_counts(normalize=True).to_dict()
    fraction_fit = fit_feedback_ratios.get('fit', 0.0)
    fraction_small = fit_feedback_ratios.get('small', 0.0)
    fraction_large = fit_feedback_ratios.get('large', 0.0)

    # Active users/products (more than 5 transactions)
    active_users = (data.groupby('user_id').size() > 5).sum()
    active_products = (data.groupby('item_id').size() > 5).sum()

    # Unique users per product
    avg_unique_users_per_product = data.groupby('item_id')['user_id'].nunique().mean()

    # Compile enhanced stats
    stats = {
        'Number of Transactions': num_transactions,
        'Numober of Customers (Unique)': num_customers,
        'Number of Products (Unique)': num_products,
        'Avg Transactions per Customer': avg_transactions_per_customer,
        'Avg Transactions per Product': avg_transactions_per_product,
        'Median Transactions per Customer': median_transactions_per_customer,
        'Median Transactions per Product': median_transactions_per_product,
        'Fraction Fit': fraction_fit,
        'Fraction Small': fraction_small,
        'Fraction Large': fraction_large,
        'Number of Active Customers (>5 Transactions)': active_users,
        'Number of Active Products (>5 Transactions)': active_products,
        'Avg Unique Users per Product': avg_unique_users_per_product
    }
    return stats

modcloth_stats = compute_statistics(cleaned_dataset_modcloth_df, "ModCloth")
renttherunway_stats = compute_statistics(renttherunway_df, "RentTheRunWay")


stats_table = pd.DataFrame({
    "ModCloth": modcloth_stats,
    "RenttheRunway": renttherunway_stats 
})
print(stats_table)
```

## Visualizing the sparsity for both the datasets
#### High Sparsity: Most entries in the matrix are empty, meaning users interact with only a small fraction of items.

#### Low Sparsity: Users interact with many items, resulting in a denser matrix.

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_sparsity(dataset, dataset_name):
    # Compute basic stats
    num_users = dataset['user_id'].nunique()
    num_items = dataset['item_id'].nunique()
    num_transactions = len(dataset)
    sparsity = 1 - (num_transactions / (num_users * num_items))
    
    print(f"Sparsity for {dataset_name}: {sparsity:.4f}")

    # Histogram of interactions per user
    interactions_per_user = dataset.groupby('user_id')['item_id'].count()
    plt.figure(figsize=(8, 6))
    plt.hist(interactions_per_user, bins=50, edgecolor='k', alpha=0.7)
    plt.title(f'Histogram of Interactions per User - {dataset_name}')
    plt.xlabel('Number of Interactions')
    plt.ylabel('Frequency')
    plt.show()

    # Histogram of interactions per item
    interactions_per_item = dataset.groupby('item_id')['user_id'].count()
    plt.figure(figsize=(8, 6))
    plt.hist(interactions_per_item, bins=50, edgecolor='k', alpha=0.7)
    plt.title(f'Histogram of Interactions per Item - {dataset_name}')
    plt.xlabel('Number of Interactions')
    plt.ylabel('Frequency')
    plt.show()

    # Heatmap of a random subset of user-item matrix
    sample_users = dataset['user_id'].drop_duplicates().sample(100).tolist()
    sample_items = dataset['item_id'].drop_duplicates().sample(100).tolist()
    sampled_matrix = dataset[
        (dataset['user_id'].isin(sample_users)) & 
        (dataset['item_id'].isin(sample_items))
    ]
    
    interaction_matrix = pd.crosstab(sampled_matrix['user_id'], sampled_matrix['item_id'])
    plt.figure(figsize=(10, 8))
    plt.imshow(interaction_matrix.values, cmap='viridis', aspect='auto')
    plt.colorbar(label='Interaction (1 = exists, 0 = no interaction)')
    plt.title(f'Interaction Heatmap - {dataset_name} (Sampled Users/Items)')
    plt.xlabel('Items')
    plt.ylabel('Users')
    plt.show()

# Visualize sparsity for RentTheRunway dataset
visualize_sparsity(renttherunway_df, "RentTheRunway")

# Visualize sparsity for ModCloth dataset
visualize_sparsity(modcloth_df, "ModCloth")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.sparse as sp
import numpy as np

def visualize_sparsity_spy(dataset, dataset_name, max_users=1000, max_items=1000):
    # Map user_id and item_id to integer indices for matrix representation
    user_mapping = {user: idx for idx, user in enumerate(dataset['user_id'].unique())}
    item_mapping = {item: idx for idx, item in enumerate(dataset['item_id'].unique())}

    # Create indices
    user_indices = dataset['user_id'].map(user_mapping)
    item_indices = dataset['item_id'].map(item_mapping)

    # Create sparse interaction matrix
    interaction_matrix = sp.coo_matrix((np.ones(len(dataset)), (user_indices, item_indices)))

    # Convert to CSR format for slicing
    interaction_matrix_csr = interaction_matrix.tocsr()

    # Subsample the data for plotting (optional for large datasets)
    max_users = min(max_users, interaction_matrix_csr.shape[0])
    max_items = min(max_items, interaction_matrix_csr.shape[1])
    sub_interaction_matrix = interaction_matrix_csr[:max_users, :max_items]

    # Plot the spy plot
    plt.figure(figsize=(8, 8))  # Square aspect ratio
    plt.spy(sub_interaction_matrix, markersize=0.5, color='black')
    plt.title(f"Sparsity Spy Plot - {dataset_name}")
    plt.xlabel(f"Items (First {max_items})")
    plt.ylabel(f"Users (First {max_users})")
    plt.gca().set_aspect('equal', adjustable='box')  # Force square aspect ratio
    plt.show()

# Visualize with controlled scales
visualize_sparsity_spy(renttherunway_df, "RentTheRunWay", max_users=1000, max_items=1000)
visualize_sparsity_spy(modcloth_df, "ModCloth", max_users=1000, max_items=1000)
```

```python
def analyze_ratings(dataset, dataset_name):
    print(f"--- Ratings Analysis for {dataset_name} ---")

    # Overall Rating Distribution
    rating_distribution = dataset['rating'].value_counts(normalize=True) * 100
    print("\nRating Distribution:")
    print(rating_distribution)

    # Ratings by Category
    if 'category' in dataset.columns:
        ratings_by_category = dataset.groupby('category')['rating'].mean().sort_values(ascending=False)
        print("\nAverage Ratings by Category:")
        print(ratings_by_category.head(10))

    # Ratings per User
    ratings_per_user = dataset.groupby('user_id')['rating'].mean()
    print("\nAverage Ratings per User (Sample):")
    print(ratings_per_user.sample(5))

# Call the function for both datasets
analyze_ratings(renttherunway_df, "RentTheRunWay")
# analyze_ratings(cleaned_dataset_modcloth_df, "ModCloth")
```

```python
import matplotlib.pyplot as plt

def plot_rating_analysis(dataset, dataset_name):
    print(f"--- Ratings Analysis for {dataset_name} ---")

    # Overall Rating Distribution
    rating_distribution = dataset['rating'].value_counts(normalize=True) * 100
    print("\nRating Distribution:")
    print(rating_distribution)

    # Plot Rating Distribution
    plt.figure(figsize=(8, 6))
    rating_distribution.sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f"Rating Distribution - {dataset_name}")
    plt.xlabel("Rating")
    plt.ylabel("Percentage")
    plt.xticks(rotation=0)
    plt.show()

    # Ratings by Category
    if 'category' in dataset.columns:
        ratings_by_category = dataset.groupby('category')['rating'].mean().sort_values(ascending=False)
        print("\nAverage Ratings by Category:")
        print(ratings_by_category.head(10))

        # Plot Top 10 Categories by Average Rating
        plt.figure(figsize=(10, 6))
        ratings_by_category.head(10).plot(kind='bar', color='coral', edgecolor='black')
        plt.title(f"Top 10 Categories by Average Rating - {dataset_name}")
        plt.xlabel("Category")
        plt.ylabel("Average Rating")
        plt.xticks(rotation=45, ha='right')
        plt.show()

    # Ratings per User
    ratings_per_user = dataset.groupby('user_id')['rating'].mean()
    print("\nAverage Ratings per User (Sample):")
    print(ratings_per_user.sample(5))

    # Plot Histogram of Average Ratings Per User
    plt.figure(figsize=(8, 6))
    ratings_per_user.plot(kind='hist', bins=20, color='green', edgecolor='black', alpha=0.7)
    plt.title(f"Distribution of Average Ratings per User - {dataset_name}")
    plt.xlabel("Average Rating")
    plt.ylabel("Frequency")
    plt.show()

# Call the function with plots for both datasets
plot_rating_analysis(renttherunway_df, "RentTheRunWay")
# plot_rating_analysis(modcloth, "ModCloth")
```

```python
def analyze_rating_trends(dataset, dataset_name):
    if 'review_date' in dataset.columns:
        print(f"--- Rating Trends for {dataset_name} ---")
        dataset['review_date'] = pd.to_datetime(dataset['review_date'])
        trends = dataset.groupby(dataset['review_date'].dt.to_period('M'))['rating'].mean()
        trends.plot(kind='line', figsize=(10, 6), title=f"Rating Trends Over Time - {dataset_name}")
        plt.xlabel("Month")
        plt.ylabel("Average Rating")
        plt.show()

# Call the function for both datasets
analyze_rating_trends(renttherunway_df, "RentTheRunWay")
# analyze_rating_trends(modcloth_df, "ModCloth")
```

```python
from textblob import TextBlob

def analyze_sentiment(dataset, dataset_name):
    print(f"--- Sentiment Analysis for {dataset_name} ---")
    if 'review_text' in dataset.columns:
        dataset['review_sentiment'] = dataset['review_text'].apply(lambda x: TextBlob(x).sentiment.polarity)
        average_sentiment = dataset['review_sentiment'].mean()
        print(f"Average Sentiment Polarity: {average_sentiment:.2f}")

    # Sentiment Comparison with Numerical Ratings
    if 'rating' in dataset.columns and 'review_sentiment' in dataset.columns:
        correlation = dataset[['rating', 'review_sentiment']].corr().iloc[0, 1]
        print(f"Correlation between Ratings and Sentiment: {correlation:.2f}")

# Call the function for both datasets
analyze_sentiment(renttherunway_df, "RentTheRunWay")
analyze_sentiment(modcloth_df, "ModCloth")
```
